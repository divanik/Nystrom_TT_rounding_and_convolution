```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
repo_path = '/home/divanik/CourseWork/Nystrom_TT_rounding_and_convolution'
```

```python
answer = []
with open('luls2.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        d, kappa, rank, precision = line.split()
        d = int(d)
        rank = int(rank)
        kappa = float(kappa)
        precision = float(precision)
        print(d, rank, kappa, precision)
        answer.append({
            'dimension' : d,
            'rank' : rank,
            'kappa' : kappa,
            'precision' : precision
        })
```

```python
pd.DataFrame(answer).to_csv(f'{repo_path}/data/experiment2/kappa.csv', index=False)
```

```python
df = pd.read_csv(f'{repo_path}/data/experiment2/kappa.csv')
for dimension in [3, 4]:
    sel_df = df[df["dimension"] == dimension]
    sns.set(font_scale=2)
    fig, ax = plt.subplots()
    fig.set_size_inches((18, 14))
    sns.lineplot(data=sel_df, x = 'rank', y = 'precision', ax = ax, hue="kappa")
    ax.set_title(f" : {dimension}")
    ax.set_yscale('log')
    ax.set_xlabel(' ')
    ax.set_ylabel(' ')
    plt.show()
```
