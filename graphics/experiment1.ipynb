```python
import pandas as pd
import seaborn as sns
```

```python
repo_dir = '/home/divanik/CourseWork/'
```

```python
import matplotlib.pyplot as plt
```

```python
kek = pd.read_csv('/home/divanik/CourseWork/Nystrom_TT_rounding_and_convolution/data/experiment2/rounding_rank.csv')
kek = kek[kek['Потоки'] == 1]
sns.set(font_scale=2)
fig, ax = plt.subplots()
fig.set_size_inches((18, 14))
sns.lineplot(data=kek, x = 'rounding_rank', y = 'time', ax = ax)
ax.set_xlabel('Ранг округления')
ax.set_ylabel('Время работы')
```

```python
kek = pd.read_csv(f'{repo_dir}/Nystrom_TT_rounding_and_convolution/data/experiment2/initial_rank.csv')
kek = kek[kek['Потоки'] == 1]
sns.set(font_scale=2)
kek['normalized_time'] = kek['time'] / (kek['initial_rank'] ** 4)
fig, ax = plt.subplots()
fig.set_size_inches((18, 14))
sns.lineplot(data=kek, x = 'initial_rank', y = 'normalized_time', ax = ax)
ax.set_xlabel('Исходный ранг')
ax.set_ylabel('Время работы / (исходный ранг)^4')
```
