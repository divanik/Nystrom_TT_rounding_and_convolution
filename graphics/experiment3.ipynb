```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
repo_path = '/home/divanik/CourseWork/Nystrom_TT_rounding_and_convolution'
```

```python
kek = pd.read_csv(f'{repo_path}/data/experiment3/kappa_dependence.csv')
sns.set(font_scale=2)
fig, ax = plt.subplots()
fig.set_size_inches((18, 14))
sns.lineplot(data=kek, x = 'rounding_rank', y = 'precision', ax = ax, hue="kappa")
ax.set_yscale('log')
ax.set_xlabel('Ранг округления')
ax.set_ylabel('Точность приближения')
```

```python
input_file = open("luls.txt", "r")

df = []

rank = None
over = None
precision = None

for line in input_file.readlines():
    if line.startswith('Rounding_rank:'):
        rank = int(line.split()[1])
    elif line.startswith('Oversampling:'):
        over = int(line.split()[1])
    elif line.startswith('Retry:'):
        pass
    else:
        precision = float(line[:-1])
        df.append({'precision' : precision, 'rounding_rank' : rank, 'oversampling' : over})
```

```python
pd.DataFrame(df).to_csv(f'{repo_path}/data/experiment3/oversampling_dependence.csv')
```
