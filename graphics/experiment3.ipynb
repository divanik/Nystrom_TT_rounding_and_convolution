```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
repo_path = '/home/divanik/CourseWork/Nystrom_TT_rounding_and_convolution'
```

```python
kek = pd.read_csv(f'{repo_path}/data/experiment3/comparison.csv')
sns.set(font_scale=2)
fig, ax = plt.subplots()
fig.set_size_inches((18, 14))
kek = kek[kek['name'] != 'tsrh(kappa=1)']
kek = kek.replace('tsrh(kappa=1.5)', 'tsrh')
sns.scatterplot(data=kek, x = 'time', y = 'precision', hue='name', ax = ax)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('Время работы')
ax.set_ylabel('Точность приближения')
```
